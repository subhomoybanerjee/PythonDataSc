```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
y
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Haha')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')


plt.subplot(1,2,2)
plt.plot(y,x,'b')

```

```python
fig=plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y,'r')
axes1.set_xlabel('X Label')
axes1.set_xlabel('Y Label')
axes1.set_title('func')

axes2=fig.add_axes([1,1,1,1])
axes2.plot(y,x,'b')
axes2.set_xlabel('X Label')
axes2.set_xlabel('Y Label')
axes2.set_title('invfunc')
```

```python
fig,axs=plt.subplots(nrows=1,ncols=2)
```

```python
for i in axs:
    i.plot(y,x)
fig
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(5,5))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
    
```

```python
fig.savefig('yolo.jpg')
```

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='square')
axes.plot(x,x**3,label='cube')

axes.legend(loc=4)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,
        color='#FF8C00',
        linewidth=3,
        alpha=0.5,
        linestyle='dotted',
        marker='o',
        markersize=20,
        markerfacecolor='yellow',
        markeredgecolor='red')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.plot(x,y)
ax2=fig.add_axes([0.2,0.5,0.2,0.2])
ax2.plot(x,y)
```

```python
fig=plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')

ax2=fig.add_axes([0.2,0.5,0.4,0.4])
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
from IPython.core.pylabtools import figsize
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,2))

axes[0].plot(x,y,color='b')
axes[1].plot(x,z,color='r',linestyle='dotted',lw=5)
```
